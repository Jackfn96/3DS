```python
import numpy as np
import math
import os
import csv
import glob
import cv2
import tensorflow as tf
import h5py as h5py
from scipy.spatial import distance
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.inception_v3 import InceptionV3, preprocess_input
from tensorflow.keras.models import Model, load_model
```

```python
def eye_aspect_ratio(eye):
    A = distance.euclidean(eye[1], eye[5])
    B = distance.euclidean(eye[2], eye[4])
    C = distance.euclidean(eye[0], eye[3])
    ear = (A + B) / (2.0 * C)
    return ear

def mouth_aspect_ratio(mouth):
    A = distance.euclidean(mouth[14], mouth[18])
    C = distance.euclidean(mouth[12], mouth[16])
    mar = (A ) / (C)
    return mar

def circularity(eye):
    A = distance.euclidean(eye[1], eye[4])
    radius  = A/2.0
    Area = math.pi * (radius ** 2)
    p = 0
    p += distance.euclidean(eye[0], eye[1])
    p += distance.euclidean(eye[1], eye[2])
    p += distance.euclidean(eye[2], eye[3])
    p += distance.euclidean(eye[3], eye[4])
    p += distance.euclidean(eye[4], eye[5])
    p += distance.euclidean(eye[5], eye[0])
    return 4 * math.pi * Area /(p**2)

def mouth_over_eye(eye):
    ear = eye_aspect_ratio(eye)
    mar = mouth_aspect_ratio(eye)
    mouth_eye = mar/ear
    return mouth_eye
```

```python
# main program
data_dir = '/Users/zohre/code/Jackfn96/sample_dataset/with_annotation/face_resize/'
faces = [f for f in os.listdir(data_dir) if '2021-10-28' in f]
print(len(faces))
print("")
faces.sort()
count = 0
data = []
labels = []
for face in faces:
    image = cv2.imread(os.path.join(data_dir, face))
    landmarks = extract_face_landmarks(image)
    if landmarks is not None:
        print('landmarks was detected.')
        count += 1
        data.append(landmarks)
        labels.append(face[:-4])
        print(count)
#     else:  
#         print("landmarks not detected, moving to the next frame ...")
```

```python
import numpy as np
data = np.array(data)
labels = np.array(labels)
```

```python
data = np.array(data)
labels = np.array(labels)
features = []
for d in data:
    eye = d[36:68] # landmarks related to eyes and ears
    ear = eye_aspect_ratio(eye)
    mar = mouth_aspect_ratio(eye)
    cir = circularity(eye)
    mouth_eye = mouth_over_eye(eye)
    features.append([ear, mar, cir, mouth_eye])

features = np.array(features)
features.shape
```
