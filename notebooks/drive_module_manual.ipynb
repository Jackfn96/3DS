You can list all files of given type

Condition must comply with Google Drive syntax https://developers.google.com/drive/api/v3/ref-search-terms

```python
drive.list('txt', add_query="name contains '26_11'")
```

## Download file

You can use `id` to download files

Returns string which can be loaded directly to pd.read_csv().

If `return bytes` is set to True, it returns byte-files. Use it for non-text file types (i.e. videos or excels)

```python
file = drive.download('18A7Ku-yd6Oi8yZj7u89ff4D_88BHp2ia')
```

```python
pd.read_csv(file)
```

## Save file

The file can be saved locally in `data` folder

```python
drive.save_locally(file, 'text.csv')
```

Doesn't matter if file was downloaded in binary or text form

```python
file_binary = drive.download('18A7Ku-yd6Oi8yZj7u89ff4D_88BHp2ia', return_bytes=True)
```

```python
drive.save_locally(file_binary, 'text2.csv')
```
